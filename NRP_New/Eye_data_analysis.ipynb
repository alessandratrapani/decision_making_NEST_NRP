# Eye joint data analysis

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

df = pd.read_csv("all_joints_positions.csv")
df.head(50)
```

```python
# Up/down eye movement

eye_tilt = df[df["Name"] == "eye_tilt"]

# Max/min extraction
max_eye_tilt = max(eye_tilt['Position'])
min_eye_tilt = min(eye_tilt['Position'])
print("Max eye tilt:", max_eye_tilt)
print("Min eye tilt:", min_eye_tilt)

fig = plt.figure(figsize=(10, 8), dpi=80)

end = eye_tilt[eye_tilt.columns[0]].count()
first_time_value = eye_tilt.iloc[0]['time']
last_time_value = eye_tilt.iloc[end-1]['time']

x = np.arange(first_time_value,last_time_value,0.02)
y = 0*x

plt.plot(eye_tilt['time'], eye_tilt['Position'], label = 'Eye tilt angle value', linewidth=2.5)
plt.plot(x, y, "-", label = '0-baseline')
plt.title('Eye tilt')
plt.ylabel('Che unità di misura è?')
plt.xlabel('Time [s]')
plt.grid(axis = 'x')
plt.legend()

eye_tilt
```

```python
# Left/right eye movement

eye_version = df[df["Name"] == "eye_version"]

# Max/min extraction
max_eye_version = max(eye_version['Position'])
min_eye_version = min(eye_version['Position'])
print("Max eye version:", max_eye_version)
print("Min eye version:", min_eye_version)

fig = plt.figure(figsize=(10, 8), dpi=80)

end = eye_version[eye_version.columns[0]].count()
first_time_value = eye_version.iloc[0]['time']
last_time_value = eye_version.iloc[end-1]['time']

x = np.arange(first_time_value,last_time_value,0.02)
y = 0*x

plt.plot(eye_version['time'], eye_version['Position'], label = 'Eye version', linewidth=2.5)
plt.title('Eye version')
plt.plot(x, y, "-", label = '0-baseline')
plt.ylabel('Che unità di misura è?')
plt.grid(axis = 'x')
plt.xlabel('Time [s]')
plt.legend()

eye_version
```

```python
# A vergence is the simultaneous movement of both eyes in opposite directions to obtain or maintain single binocular vision.
# It can be either convergence or divergence of the eyes

eye_vergence = df[df["Name"] == "eye_vergence"]

# Max/min extraction
max_eye_vergence = max(eye_vergence['Position'])
min_eye_vergence = min(eye_vergence['Position'])
print("Max eye vergence:", max_eye_vergence)
print("Min eye vergence:", min_eye_vergence)

fig = plt.figure(figsize=(10, 8), dpi=80)

end = eye_vergence[eye_vergence.columns[0]].count()
first_time_value = eye_vergence.iloc[0]['time']
last_time_value = eye_vergence.iloc[end-1]['time']

x = np.arange(first_time_value,last_time_value,0.02)
y = 0*x;

plt.plot(eye_vergence['time'], eye_vergence['Position'], label = 'Eye vergence', linewidth=2.5)
plt.title('Eye vergence')
plt.plot(x, y, "-", label = '0-baseline')
plt.ylabel('Che unità di misura è?')
plt.grid(axis = 'x')
plt.xlabel('Time [s]')
plt.legend()

eye_vergence
```
